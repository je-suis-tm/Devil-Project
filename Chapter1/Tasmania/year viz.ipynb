```python
sim_years=40
gridlen=35
```

```python
df=pd.read_csv('raw.csv')
weights=pd.read_csv('weights.csv')
land=pd.read_csv('land.csv')
land['landarea']-=1
landlength=len(land)
rates=pd.read_csv('rates.csv')
```

```python
del df['Unnamed: 0']

arr=np.array(df)

coeffs=np.repeat(np.array(weights['weight']).reshape(-1,1),landlength,axis=1)

landmap=np.array(land['landarea'])

yearmap=[]
for year in range(sim_years):
    singleyear=arr[:,year*landlength:(year+1)*landlength]
    yearmap.append(np.multiply(singleyear,coeffs).sum(axis=0))
```

```python
for i in range(40):
    
    hs=np.full(gridlen**2, np.nan)

    hs[landmap]=yearmap[i]
    hs=hs.reshape(gridlen,gridlen)

    plt.imshow(hs,cmap='RdBu_r',vmin=hs[~np.isnan(hs)].min(),
           vmax=hs[~np.isnan(hs)].max())
    
    plt.title(f'year {i}')
    plt.axis('off')
    plt.show()
```

```python
for i in range(40):
    
    hs=np.full(gridlen**2, np.nan)

    hs[landmap]=yearmap[i]
    hs=hs.reshape(gridlen,gridlen)

    plt.imshow(hs,cmap='RdBu_r',vmin=np.array(yearmap).min(),
           vmax=np.array(yearmap).max())
    
    plt.title(f'year {i}')
    plt.axis('off')
    plt.show()
```
